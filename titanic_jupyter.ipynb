```python
target_variable = 'Survived'
```

```python
plotted_variables = ['PassengerId', 'Pclass','SibSp','Parch','Age','Fare']
```

```python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize = (15,10))

for i, var in enumerate(plotted_variables):
    row, col = divmod(i, 3)  # Calculate the row and column for the subplot
    ax = axes[row, col]
    
    ax.scatter(df_train[var], df_train[target_variable], alpha=0.5)  # Create scatter plot
    ax.set_xlabel(var)  # Set x-axis label
    ax.set_ylabel(target_variable)  # Set y-axis label

    # Adjust layout and display the plots
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns


# Define the categorical features to explore
categorical_features = ['Pclass', 'Sex', 'Embarked']

# Create subplots for each categorical feature
fig, axes = plt.subplots(1, len(categorical_features), figsize=(15, 5))

# Iterate through categorical features and create bar plots
for i, feature in enumerate(categorical_features):
    ax = axes[i]
    sns.barplot(x=feature, y='Survived', data=df_train, ax=ax, ci=None)
    ax.set_title(f'Survival Rate by {feature}')
    ax.set_ylabel('Survival Rate')
    ax.set_xlabel(feature)

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```

```python
feature1 = 'Survived'
feature2 = 'Fare'

# Create a scatter plot using Seaborn
sns.set(style='whitegrid')  # Optional: Set a seaborn style
plt.figure(figsize=(8, 6))
sns.scatterplot(x=feature1, y=feature2, data=df_train)
plt.title(f'Scatter Plot: {feature1} vs. {feature2}')
plt.xlabel(feature1)
plt.ylabel(feature2)

# Alternative: Create a scatter plot using Matplotlib
# plt.figure(figsize=(8, 6))
# plt.scatter(df[feature1], df[feature2], alpha=0.5)
# plt.title(f'Scatter Plot: {feature1} vs. {feature2}')
# plt.xlabel(feature1)
# plt.ylabel(feature2)

plt.show()
```
